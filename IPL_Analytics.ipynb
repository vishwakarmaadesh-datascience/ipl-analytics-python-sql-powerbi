```python
import pandas as pd
import numpy as np
import os
os.listdir()
```

```python
datasets = {}
for file in os.listdir():
    if file.endswith(".csv"):
        datasets[file] = pd.read_csv(file)

print(datasets.keys())
```

```python
for name, df in datasets.items():
    print("="*50)
    print("Dataset:", name)
    print("Shape:", df.shape)
    print(df.head())
    print("\n")
```

```python
all_ipl_data = pd.concat(datasets.values(), ignore_index=True)

print("Shape of the combined dataset:", all_ipl_data.shape)
display(all_ipl_data.head())
```

```python
df.columns = df.columns.str.lower().str.strip()
df.to_csv("cleaned_matches.csv", index=False)
```

### 4. Data Cleaning

Given the high percentage of missing values in certain columns (e.g., `year`, `event_name`, `match_number`, `match_type`, `balls_per_over`, `gender`, `team_type`),  drop these columns as they might not provide substantial information for all records. For columns like `winner` and `player_of_match`, which have fewer missing values, we will fill them with 'Unknown' as they are categorical.

```python
df['date'] = pd.to_datetime(df['date'])
df.info()
```

```python
columns_to_drop = ['year', 'event_name', 'match_number', 'match_type', 'balls_per_over', 'gender', 'team_type']
all_ipl_data = all_ipl_data.drop(columns=columns_to_drop, errors='ignore')

initial_rows = all_ipl_data.shape[0]
all_ipl_data.drop_duplicates(subset=['date', 'team1', 'team2'], inplace=True)
duplicates_removed = initial_rows - all_ipl_data.shape[0]
print(f"Removed {duplicates_removed} duplicate match entries.")

# Fill missing values in 'winner' and 'player_of_match' with 'Unknown'
all_ipl_data['winner'] = all_ipl_data['winner'].fillna('Unknown')
all_ipl_data['player_of_match'] = all_ipl_data['player_of_match'].fillna('Unknown')

print("After cleaning, missing values percentage per column:")
display(all_ipl_data.isnull().sum() / len(all_ipl_data) * 100)
print("Shape of the dataset after dropping columns and duplicates:", all_ipl_data.shape)
```

### 5. Standardizing Team Names

Some IPL teams have changed their names over the seasons. To ensure consistency and facilitate analysis, we will standardize these team names to their most recent or commonly recognized names.

```python
team_name_map = {
    'Delhi Daredevils': 'Delhi Capitals',
    'Kings XI Punjab': 'Punjab Kings',
    'Rising Pune Supergiant': 'Rising Pune Supergiants',
    'Royal Challengers Bangalore': 'Royal Challengers Bengaluru',
    # add any other variants you found in the unique() output
}

all_ipl_data['team1'] = all_ipl_data['team1'].replace(team_name_map)
all_ipl_data['team2'] = all_ipl_data['team2'].replace(team_name_map)
all_ipl_data['winner'] = all_ipl_data['winner'].replace(team_name_map)
all_ipl_data['toss_winner'] = all_ipl_data['toss_winner'].replace(team_name_map)

# Verify
all_teams = pd.concat([all_ipl_data['team1'], all_ipl_data['team2']]).unique()
print(sorted(all_teams))
```

### 4. Unique Values in Key Categorical Columns

Let's examine the unique values and their counts for important categorical features like `season`, `city`, `venue`, `toss_decision`, and `winner`.

```python
categorical_cols = ['season', 'city', 'venue', 'toss_decision', 'winner', 'team1', 'team2', 'player_of_match']

for col in categorical_cols:
    if col in all_ipl_data.columns:
        print(f"\nUnique values in '{col}': {all_ipl_data[col].nunique()}")
        display(all_ipl_data[col].value_counts().head())
        print("-"*30)
```

### 6. Relationship Between Toss Decision and Match Outcome

Let's see if winning the toss and choosing to bat or field has an impact on the match outcome.

```python
toss_winner_match_winner = all_ipl_data[all_ipl_data['toss_winner'] == all_ipl_data['winner']]
toss_winner_match_loser = all_ipl_data[all_ipl_data['toss_winner'] != all_ipl_data['winner']]

win_after_toss_win_percentage = len(toss_winner_match_winner) / len(all_ipl_data) * 100

print(f"Percentage of matches where toss winner also won the match: {win_after_toss_win_percentage:.2f}%")

plt.figure(figsize=(10, 6))
sns.countplot(data=all_ipl_data, x='toss_decision', hue='winner', palette='viridis')
plt.title('Match Outcome based on Toss Decision')
plt.xlabel('Toss Decision')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Match Winner')
plt.tight_layout()
plt.show()
```

### 7. Match Results by Win Margin

Let's visualize the distribution of wins by runs and wickets.

```python
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
sns.histplot(all_ipl_data['win_by_runs'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Wins by Runs')
plt.xlabel('Runs')
plt.ylabel('Number of Matches')

plt.subplot(1, 2, 2)
sns.histplot(all_ipl_data['win_by_wickets'], bins=10, kde=True, color='lightcoral')
plt.title('Distribution of Wins by Wickets')
plt.xlabel('Wickets')
plt.ylabel('Number of Matches')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
sns.boxplot(y=all_ipl_data['win_by_runs'], color='skyblue')
plt.title('Box Plot of Wins by Runs')
plt.ylabel('Runs')

plt.subplot(1, 2, 2)
sns.boxplot(y=all_ipl_data['win_by_wickets'], color='lightcoral')
plt.title('Box Plot of Wins by Wickets')
plt.ylabel('Wickets')

plt.tight_layout()
plt.show()
```

```python
total_matches = len(all_ipl_data)

# Matches won by runs (win_by_runs > 0)
matches_won_by_runs = all_ipl_data[all_ipl_data['win_by_runs'] > 0]
percentage_won_by_runs = (len(matches_won_by_runs) / total_matches) * 100

# Matches won by wickets (win_by_wickets > 0)
matches_won_by_wickets = all_ipl_data[all_ipl_data['win_by_wickets'] > 0]
percentage_won_by_wickets = (len(matches_won_by_wickets) / total_matches) * 100

print(f"Percentage of matches won by runs: {percentage_won_by_runs:.2f}%")
print(f"Percentage of matches won by wickets: {percentage_won_by_wickets:.2f}%")
```

```python
# Convert 'season' to numeric for proper sorting in plots
all_ipl_data['season'] = pd.to_numeric(all_ipl_data['season'], errors='coerce')
all_ipl_data.dropna(subset=['season'], inplace=True)
all_ipl_data['season'] = all_ipl_data['season'].astype(int)

# Calculate wins by runs and wickets per season
season_wins = all_ipl_data.groupby('season').agg(
    total_matches=('season', 'count'),
    wins_by_runs=('win_by_runs', lambda x: (x > 0).sum()),
    wins_by_wickets=('win_by_wickets', lambda x: (x > 0).sum())
).reset_index()

# Calculate percentages
season_wins['percentage_wins_by_runs'] = (season_wins['wins_by_runs'] / season_wins['total_matches']) * 100
season_wins['percentage_wins_by_wickets'] = (season_wins['wins_by_wickets'] / season_wins['total_matches']) * 100

print("Win percentages by season:")
display(season_wins[['season', 'percentage_wins_by_runs', 'percentage_wins_by_wickets']].head())
```

### Win Percentages by Season

```python
plt.figure(figsize=(14, 7))

# Melt the DataFrame for easier plotting with seaborn
season_wins_melted = season_wins.melt(id_vars=['season'],
                                      value_vars=['percentage_wins_by_runs', 'percentage_wins_by_wickets'],
                                      var_name='Win Type',
                                      value_name='Percentage')

sns.barplot(x='season', y='Percentage', hue='Win Type', data=season_wins_melted, palette='viridis')
plt.title('Percentage of Matches Won by Runs vs. Wickets Across Seasons')
plt.xlabel('Season')
plt.ylabel('Percentage of Matches Won')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Win Type')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
